```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv('D:\\Y3S1 DS\\IRWA\\Product_Categorization_AI\\backend\\data\\product.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nDataset Info:")
print(df.info())
```

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Check data types
print("\nData Types:")
print(df.dtypes)

# Check unique values for each column
print("\nUnique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Check which columns have missing values
missing_cols = df.columns[df.isnull().any()].tolist()
print("Columns with missing values:", missing_cols)

# Fill missing values appropriately
# For categorical columns, fill with mode
# For numerical columns, fill with median or appropriate value

# Example: If 'season' has missing values
if 'season' in df.columns and df['season'].isnull().any():
    df['season'].fillna(df['season'].mode()[0], inplace=True)

# Check if any missing values remain
print("Missing values after treatment:", df.isnull().sum().sum())
```

```python
# Convert appropriate columns to categorical
categorical_cols = ['gender', 'masterCategory', 'subCategory', 'articleType', 
                   'baseColour', 'season', 'usage']

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Convert year to datetime or keep as numeric based on need
if 'year' in df.columns:
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

```python
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

```python
# Create new features if needed
# Example: Extract brand name from productDisplayName
df['brand'] = df['productDisplayName'].apply(lambda x: x.split()[0] if pd.notnull(x) else 'Unknown')

# Example: Create length of product name feature
df['name_length'] = df['productDisplayName'].apply(lambda x: len(str(x)))
```

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if numerical_cols:
    print("Numerical columns:", numerical_cols)
    
    # Visualize outliers using boxplots
    plt.figure(figsize=(15, 5))
    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(1, len(numerical_cols), i)
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Label encoding for categorical variables
label_encoders = {}
categorical_cols = df.select_dtypes(include=['category', 'object']).columns.tolist()

for col in categorical_cols:
    if col not in ['productDisplayName', 'filename', 'link']:  # Skip text columns
        le = LabelEncoder()
        df[col + '_encoded'] = le.fit_transform(df[col].astype(str))
        label_encoders[col] = le

print("Categorical columns encoded")
```

```python
# Final dataset overview
print("Final Dataset Info:")
print(df.info())
print("\nFinal Dataset Shape:", df.shape)
print("\nMissing Values Summary:")
print(df.isnull().sum())

# Basic statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Save the cleaned dataset
df.to_csv('cleaned_product_data.csv', index=False)
print("Cleaned dataset saved as 'cleaned_product_data.csv'")
```
